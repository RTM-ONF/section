```python
from matplotlib import pyplot as plt
import pandas as pd

from section import Section
```

```python
df = pd.read_csv("../dataset/section_bis.txt",
                 delimiter="\t",
                 decimal=".")
```

```python
section = Section()
section.from_df(df, x_field="X", z_field="Z")
```

```python
section.xz
```

```python


fig, ax = plt.subplots()
ax.set_aspect("equal")
ax.plot(section.x, section.z)
for line in section.water_lines(altitude=548):
    ax.plot([x for x, z in line], [z for x, z in line], color="Red")
plt.show()
```

```python
min(section.z)
```

```python
# intersection avec point le plus bas
section.intersect([(section.x[0], min(section.z)),(section.x[-1], min(section.z))])
```
